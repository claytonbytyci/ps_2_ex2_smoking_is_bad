```python
# TODO: install all the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# TODO: load the data using the pathlib library to create the path to your data

# Create a Path object pointing to your data file
data_path = Path("data") / "smoking_data.csv"  # adjust folder/filename as needed

# Load the data into a pandas DataFrame
df = pd.read_csv(data_path)
```

```python
# TODO: what are the dimensions of the data frame?
df.shape
```

```python
# TODO: use .describe() on your data frame: what do you notice about the string columns?
df.describe(include='all')
```

```python
# TODO: do not forget to check the null values
df.isnull().sum()
```

```python
# TODO: prepare your data frame so that you can work with

# Drop the 'Unnamed: 0' column
df = df.drop(columns=['Unnamed: 0'])

# Turn 'gender' into a binary 'male' column (1 if male, 0 if female)
df['male'] = df['gender'].apply(lambda x: 1 if x.lower() == 'male' else 0)

# Turn 'outcome' into a binary 'alive' column (1 if Alive, 0 if Dead)
df['alive'] = df['outcome'].apply(lambda x: 1 if x.lower() == 'alive' else 0)

# Drop rows with any missing (NaN) values
df = df.dropna()

df
```

# Is smoking really good?
Now it is up to you to run the analysis further and check what could be underlying your cousings results.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style
sns.set(style="whitegrid")

sns.boxplot(x='smoker', y='salary', data=df)
plt.title("Smokers vs Non-Smokers: Salary Distribution")
plt.show()

sns.barplot(x='smoker', y='alive', data=df, ci=None)
plt.title("Proportion Alive: Smokers vs Non-Smokers")
plt.show()
```

```python
sns.boxplot(x='smoker', y='age', data=df)
plt.title("Smokers vs Non-Smokers: Age Distribution")
plt.show()
```

```python
sns.scatterplot(x='age', y='salary', hue='smoker', data=df)
plt.title("Age vs Salary by Smoking Status")
plt.show()

sns.barplot(x='smoker', y='alive', hue='male', data=df, ci=None)
plt.title("Proportion Alive: Smoker vs Non-Smoker (by Gender)")
plt.show()
```
